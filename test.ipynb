```python
import tkinter as tk

data = {"Aman": 5000, "Rohit": 3000, "Devik": 7000}

root = tk.Tk()

text = tk.Text(root, height=10, width=40)
text.pack()

for k, v in data.items():
    text.insert(tk.END, f"{k} : {v}\n")

root.mainloop()
```
